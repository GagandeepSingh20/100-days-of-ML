```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('~/Desktop/Student Depression Dataset.csv')
```

```python
df.sample(3)
```

```python
df['Dietary Habits'].unique()
df = df.dropna(subset=["Sleep Duration"])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder 
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,1:16],df.iloc[:,-1], test_size=0.2)
```

```python
trans= ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['Academic Pressure']),
    ('tnf2',OrdinalEncoder(categories=[['Others','Unhealthy','Moderate','Healthy']]),['Dietary Habits']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['Profession','Gender','City']),
],remainder='passthrough')
```
